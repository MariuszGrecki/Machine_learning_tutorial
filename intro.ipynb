```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')
df.info()
```

Standaryzacja - StandardScaler

std() - pandas nieobciązony

std() - numpy obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}") # odchylenie standardowe jest miarą rozproszenia danych wokół średniej w zbiorze danych. Jest to jedno z podstawowych narzędzi w statystyce
```

```python
# standaryzacja, takiego dane z reguły wrzucamy do modelu

(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
# równie dobrze możemy do tego celu zbudować własną funkcję

def standardize(x):
  return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
#jest też gotowa funkcja w scikit, wyniki mogą się różnić na względu estymalator obciążony / nieobciążony

from sklearn.preprocessing import scale

scale(df['price'])
```

```python
#jest też gotowa klasa, jest lepsza pozwala zachować dane

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
#mozna standaryzować naraz dwie kolumny
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```

**Przygotowanie danych do modelu**

```python
#to tylko skopiowanie pierwotnej postaci
df = df_raw.copy()
df
```

```python
#nowa klasa labelEncoder
le = LabelEncoder()

#metoda fit uczy się unikalnych wartosci w kolumnie i przypisuje im etykiety numeryczne, natomiast transform przekształca rzeczywiste wartości w kolumnie na ich odpowiadające wartości numeryczne czyli yes => 1
df['bought'] = le.fit_transform(df['bought'])

#odchylenie
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

#tutaj zamieniamy red czy female na kolumny
df = pd.get_dummies(data=df, drop_first=True)

# Zamiana wartości False/True na 0/1 dla wszystkich kolumn typu bool
for column in df.columns:
    if df[column].dtype == 'bool':
        df[column] = df[column].astype(int)

df
```
