```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
np.set_printoptions(precision=6, suppress = True, edgeitems=30, linewidth=120, formatter=dict(float=lambda x: f'{x:.2f}'))

sns.set(font_scale=1.3)
sklearn.__version__
```

```python
from sklearn.datasets import make_regression

data, target = make_regression(n_samples=100, n_features=1, n_targets=1, noise=30.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
data[:5]
```

```python
target[:5]
```

```python
plt.figure(figsize=(8, 6))
plt.title('regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.plot()
```

**Regresja liniowa przy użyciu scikit-learn**

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

#metoda fit() dopasowuje model liniowy do danych
regressor.fit(data, target)
```

```python
#metoda score() dokonuje oceny modelu na przekazanych danych (wynik R2 score)
regressor.score(data, target)
```

```python
#metoda predict() dokonuje predykcji na podstawie modelu
y_pred = regressor.predict(data)
y_pred
```

```python
#wizualizacja graficzna modelu

plt.figure(figsize=(8, 6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.plot(data, y_pred, color='red', label='model')
plt.legend()
plt.show()
```

Współczynnik R2 jest zdefiniowany jako 1 - (u/v), gdzie u jest okreslone wzorem:

((y_true - y_pred) ** 2).sum()

oraz v jako

((y_true - y_true.mean()) ** 2).sum()

Najlepszą możliwoa wartościa jest 1.0 R2 może przyjmować wartość ujemną. Model, który przeiwduje zawsze dla każdej próbki wartośc oczekiwaną zmiennej docelowej ma współczynnik R2 równy 0.

**Regresja z podziałem na zbiór treningowy oraz testowy**


```python
#wykonujemy 1000 próbek
data, target = make_regression(n_samples=1000, n_features=1, n_targets=1, noise=15.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
plt.figure(figsize=(8, 6))
plt.title('regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.plot()
```

```python
#podział danych na zbiór treningowy i testowy
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
#przygotowanie wykresu z podziałem na dwa zbiory treningowy oraz testowy
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa train vs. test')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train, y_train, label='zbiór treningowy', color='gray', alpha=0.5)
plt.scatter(X_test, y_test, label='zbiór testowy', color='gold', alpha=0.5)
plt.legend()
plt.plot()
```

```python
#tworzymy instancje klasy oraz dzięki metodzie fit dopasowujemy nasze dane treningowe
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#oceniamy nasz model na danych treningowych
regressor.score(X_train, y_train)
```

```python
#oceniamy nasz model na danych testowych
regressor.score(X_test, y_test)
```

```python
#regresja liniowa - zbiór treningowy - wizualizacja
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa: zbior treningowy')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train, y_train, label='zbiór treningowy', color='gray', alpha=0.5)
plt.plot(X_train, regressor.intercept_ + regressor.coef_[0] * X_train, color='red')
plt.legend()
plt.plot()
```

```python
#regresja liniowa - zbiór testowy - wizualizacja
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa: zbiór testowy')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_test, y_test, label='zbiór testowy', color='gold', alpha=0.5)
plt.plot(X_test, regressor.intercept_ + regressor.coef_[0] * X_test, color='red')
plt.legend()
plt.plot()
```

```python
#PREDYCKJA na podstawie modelu - Predykcja na podstawie modelu w kontekście machine learning oznacza przewidywanie przyszłych wartości lub wyników
#na podstawie wzorców i zależności, które model maszynowego uczenia się nauczył się z danych treningowych
```

```python
y_pred = regressor.predict(X_test)
predictions = pd.DataFrame(data={'y_true': y_test, 'y_pred': y_pred})
predictions.head()
```

```python
#dodajemy kolejną kolumnę która policzy nam błąd predykcji
predictions['error'] = predictions['y_true'] - predictions['y_pred']
predictions.head()
```

```python
#rozkład naszego błędu
_ = predictions['error'].plot(kind='hist', bins=50, figsize=(8,6))
```
