**Zbiór treningowy, walidacyjny i testowy**

1. Podejście

Trenowanie modelu na zbiorze treningowym i tylko na tym zbiorze -> Ocena modelu na zbiorze testowym -> Dopasowanie modelu zgodnie z wynikami na zbiorze testowym
-> wybór modelu, który radzi sobie najlepiej na zbiorze testowym

2. Podejście (tutaj jest też zbiór walidacyjny)

Trenowanie modelu na zbiorze treningowym -> ocena modelu na zbiorze walidacyjnym -> dopasowanie modelu zgodnie z wynikami na zbiorze walidacyjnym -> wybór modelu, który najlepiej radzi sobie na zbiorze wadilcyjnym -> potwierdzenie wyników na zbiorze testowym

```python
import numpy as np
import pandas as pd
import sklearn
#seaborn do wizualizacji
import seaborn as sns

#ziarno losowe te same dane podczas losowania
np.random.seed(42)
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: f'{x:.2f}'))

#seaborn jako domyślny motyw
sns.set()
sklearn.__version__
```

```python
#sklearn posiada dane testowe
from sklearn.datasets import load_iris

raw_data = load_iris()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

**Przygotowanie danych**

```python
data = raw_data_copy['data']
target = raw_data_copy['target']

print(f'{data[:5]}')
print(target[:5])
```

```python
#połączenie atrybutów ze zmienną docelową

all_data = np.c_[data, target]
all_data[:5]
```

```python
#budowa obiektu DataFrame
df = pd.DataFrame(data=all_data, columns=raw_data.feature_names + ['target'])
df.head()
```

```python
#wyświetlanie 5 ostatnich wierszy metoda tail()
df.tail()
```

```python
df.info()
```

```python
#proste statystyki odnośnie naszych danych
df.describe()
```

```python
#uproszczenie wyświetlania
df.describe().T.apply(lambda x: round(x, 2))
```

```python
#możemy znaleźć rozkład zmiennej
df.target.value_counts()
```

```python
#tutaj akurat wykres kołowy do ostaniej rzeczy
df.target.value_counts().plot(kind='pie')
```

```python
data = df.copy()
#metoda pop wyrywa 'target' z data
target = data.pop('target')
data.head()
```

```python
#wyrwane dane
target.head()
```

**Podział danych na zbiór treningowy i testowy - iris data**

```python
from sklearn.model_selection import train_test_split

#zwróci nam 4 cechy
# X_train - cechy w zbiorze treningowym
# X_test - cechy w zbiorze testowym
# y_trains - zmienna treningowa w zbiorze ttreningowym
# y_test - zmienna treningowa w zbiorze testowym

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 42)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
#kontrola train_test_split dzięki innym parametrom
#test_size domyślnie 0.25?

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.3, random_state = 42)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
#kontrola train_test_split dzięki innym parametrom
#train size = udział elementów w zbiorze treningowym do testowego

X_train, X_test, y_train, y_test = train_test_split(data, target, train_size = 0.9, random_state = 42)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

```python
#kontrola train_test_split dzięki innym parametrom
#stratify - równy podział względem zmiennej docelowej

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 42, test_size=0.1, stratify=target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny train:\n{y_train.value_counts()}')
print(f'\ny_test:\n{y_test.value_counts()}')
```

**przykład użycia zbiory treningowego i testowego na breact cander data**


```python
from sklearn.datasets import load_breast_cancer

raw_data = load_breast_cancer()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

```python
data = raw_data_copy['data']
target = raw_data_copy['target']

print(f'{data[:5]}\n')
print(target[:5])
```

```python
all_data = np.c_[data, target]
all_data[:5]
```

```python
df = pd.DataFrame(data=all_data, columns=list(raw_data['feature_names']) + ['target'])
df.head()
```

```python
#duża różnica warto użyć stratify
df.target.value_counts()
```

```python
data = df.copy()
target = data.pop('target')
data.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state = 40, test_size=0.25, stratify=target)

print(f'X_train shape {X_train.shape}')
print(f'y_train shape {y_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_test shape {y_test.shape}')
print(f'\nTest ratio: {len(X_test) / len(data):.2f}')
print(f'\ny_train:\n{y_train.value_counts() / len(y_train)}')
print(f'\ny_test:\n{y_test.value_counts() / len(y_test)}')
```
