# Klasyfikator wieloklasowy - Image classification

Fashion Minst to zestaw danych z obrazkami artykułów Zalando - składający się z zestawu treningowego 60 000 przykładów i zestawu testowego 10 000 przykładów. Każdy przykład to obraz w skali szarości 28x28, powiązany z jedna etykieta z 10 klas.

image.png

```python
%tensorflow_version 2.x
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets.fashion_mnist import load_data

np.set_printoptions(precision=12, suppress=True, linewidth=150)
np.random.seed(42)
sns.set(font_scale=1.3)
```

# Załadowanie danych i wstępna eksploracja

```python
(X_train, y_train), (X_test, y_test) = load_data()
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
print(f'X_train[0] shape: {X_train[0].shape}')
```

```python
X_train[0]
```

```python
y_train[:10]
```

```python
plt.imshow(X_train[0], cmap='gray_r')
plt.axis('off')
```

```python
class_names = ['T-shirt', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(18,13))
for i in range(1,11):
  plt.subplot(1, 10, i)
  plt.axis('off')
  plt.imshow(X_train[i-1], cmap='gray_r')
  plt.title(class_names[y_train[i-1]], color='black', fontsize=16)
plt.show()
```

```python
#przygotowanie danych
X_train = X_train / 255
X_test = X_test / 255
```

```python
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

```python
#spłaszczenie danych
X_train = X_train.reshape(60000, 28 * 28)
X_test = X_test.reshape(10000, 28 * 28)

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

```python
#budowa modelu

from sklearn.svm import SVC #grid search zajmie tutaj troche czasu

classifier = SVC()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#Ocena modelu

y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

cm
```

```python

plt.figure(figsize=(8,6))
plt.title('Macierz konfuzji')
_ = sns.heatmap(cm, annot=True, cmap=sns.cm.rocket_r)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=class_names))
```

```python
#Błędnie sklasyfikowane obrazy
results = pd.DataFrame(data={'y_pred':y_pred, 'y_test': y_test})
results.head(10) # na 10 wierszy nie ma błędów
```

```python
errors = results[results['y_pred'] != results['y_test']]
errors_idxs = list(errors.index)
errors_idxs[:10]
```

```python
results.loc[errors_idxs[:10], :]
```

```python
#wyświetlamy 4 pierwsze błędy
plt.figure(figsize=(16,10))
for idx, error_idx in enumerate(errors_idxs[:15]):
  image = X_test[error_idx].reshape(28,28)
  plt.subplot(3,5, idx + 1)
  plt.axis('off')
  plt.imshow(image, cmap='Greys')
  plt.title(f"T- {class_names[results.loc[error_idx, 'y_test']]} P-: {class_names[results.loc[error_idx, 'y_pred']]}")
```
